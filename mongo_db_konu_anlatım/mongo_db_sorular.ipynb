# Soru 4

Kitabın Basım yılının 1950'den fazla olan oları yazdırın. For döngüsü ile anlaşılabilir bir format oluşturun. Bunun için '$gte' methodunu kullanın

```python
veriler  = tablo.find({'year':{'$gte':1950}})
for veri in veriler:
    print(f"""
        ---------------------------------
        Kitap Türü: {veri['genre']}

        Kitap Adı: {veri['title']}

        Kitap Yazarı: {veri['author']}

        Kitap Basım Yılı: {veri['year']}
        ----------------------------------
    """)
```
